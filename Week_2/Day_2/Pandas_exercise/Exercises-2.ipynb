```python
import pandas as pd
```

### Download the data and load it to Pandas. 

You can find them [here](https://drive.google.com/file/d/1NY6cmF9Shjw-dD7BD6bNmfcIVz-kQcFR/view?usp=sharing).

```python
titles = pd.read_csv('titles.csv', index_col=None)
titles.head()
```

```python
cast = pd.read_csv('cast.csv', index_col=None)
cast.head()
```

### What are the ten most common movie names of all time?

```python
titles.value_counts('title').head(10)
```

### Which three years in the 1930s saw the most films released?

```python
titles[titles.year // 10 == 193].value_counts('year').head(3)
```

### Count the number of films that have been released in each decade over the history of cinema.

```python
titles.title.groupby((titles.year // 10) * 10).count()
```

### Count the number of "Hamlet" films made in each decade.

```python
titles.title[titles.title == 'Hamlet'].groupby((titles.year // 10) * 10).count()
```

### Count the number of "Rustler" characters in each decade of the history of film.

```python
cast.character[cast.character == 'Rustler'].groupby((cast.year // 10) * 10).count()
```

### Count the number of "Hamlet" characters in each decade.

```python
cast.character[cast.character == 'Hamlet'].groupby((cast.year // 10) * 10).count()
```

### What are the 11 most common character names in movie history?

```python
cast.character.groupby(cast.character).count().sort_values(ascending = False).head(11)
```

### Who are the 10 people most often credited as "Herself" in film history?

```python
cast.name[cast.character == 'Herself'].groupby(cast.name).count().sort_values(ascending = False).head(10)
```

### Who are the 10 people most often credited as "Himself" in film history?

```python
cast.name[cast.character == 'Himself'].groupby(cast.name).count().sort_values(ascending = False).head(10)
```

### Which actors or actresses appeared in the most movies in the year 1945?

```python
cast.name[((cast.type == 'actor') | (cast.type == 'actress')) & (cast.year == 1945)].groupby(cast.name).count().sort_values(ascending = False).head(1)
```

### Which actors or actresses appeared in the most movies in the year 1985?

```python
cast.name[((cast.type == 'actor') | (cast.type == 'actress')) & (cast.year == 1985)].groupby(cast.name).count().sort_values(ascending = False).head(1)
```

### How many roles "Mammootty" has played in each year of his career.

```python
cast.character[cast.name == 'Mammootty'].groupby(cast.year).count()
```

### What are the 10 most frequent roles that start with the phrase "Patron in"?

```python
cast.character[cast.character.str.startswith('Patron in')].groupby(cast.character).count().sort_values(ascending = False).head(10)
```

### What are the 10 most frequent roles that start with the word "Science"?

```python
cast.character[cast.character.str.startswith('Science')].groupby(cast.character).count().sort_values(ascending = False).head(10)
```

### Count number of n-values of the roles that Judi Dench has played over her career.

```python
cast.n[cast.name == 'Judi Dench'].sum()
```

### Count number of n-values of Cary Grant's roles through his career.

```python
cast.n[cast.name == 'Cary Grant'].sum()
```

### Count number of n-value of the roles that Sidney Poitier has acted over the years.

```python
cast.n[cast.name == 'Sidney Poitier'].sum()
```

### How many leading (n=1) roles were available to actors, and how many to actresses, in the 1950s?

```python
cast.type[(cast.n == 1) & (cast.year // 10 == 195)].groupby(cast.type).count()
```

### How many supporting (n=2) roles were available to actors, and how many to actresses, in the 1950s?

```python
cast.type[(cast.n == 2) & (cast.year // 10 == 195)].groupby(cast.type).count()
```
