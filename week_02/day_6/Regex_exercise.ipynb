## Print out every line from the file that...

... that has 'q'

```python
for line in lines:
    if re.search(r'q', line) != None:
        print(line)
```

```python
# Check with Python

for line in lines:
    if 'q' in line:
        print(line)
```

... that starts with 'H'

```python
for line in lines:
    if re.search(r'H', line) != None:
        print(line)
```

```python
# Check with Python

for line in lines:
    if 'H' in line:
        print(line)
```

... that has 'wh'

```python
for line in lines:
    if re.search(r'wh', line) != None:
        print(line)
```

```python
# Check with Python

for line in lines:
    if 'wh' in line:
        print(line)
```

... that has an 'q' or a 'Q'

```python
for line in lines:
    if re.search(r'[qQ]', line):
        print(line)
```

... that has a '*' in it

```python
for line in lines:
    if re.search(r'\*', line):
        print(line)
```

... that has a '*' (star) in it

```python
for line in lines:
    if re.findall(r'\*\.', line):
        print(line)
```

... that starts with an 'T' or an 't'

```python
for line in lines:
    if re.search(r'^T|^t', line):
        print(line)
```

... that starts with number

```python
for line in lines:
    if re.search(r'^\d', line):
        print(line)
```

... that has both 'a' and 'e'  and 'i' and 'o' and 'u' in it

```python
for line in lines:
    if re.search(r'a|A', line) and re.search(r'e|E', line) and re.search(r'i|I', line) and re.search(r'o|O', line) and re.search(r'u|U', line):
        print(line)
```

... that has an 'a' and somewhere later an 'e'

```python
for line in lines:
    if re.search(r'a.*e', line):
        print(line)
```

... that does not have an 'i'

```python
for line in lines:
    if re.search(r'^((?!i).)*$', line):
        print(line)
```

```python
# Test with Python

for line in lines:
    if 'i' not in line:
        print(line)
```

... that does not have an 'i' nor 'z'

```python
for line in lines:
    if re.search(r'^((?!i).)*$', line) and re.search(r'^((?!z).)*$', line):
        print(line)
```

... that has an 'x' but not 'y'

```python
for line in lines:
    if re.search(r'x', line) and re.search(r'^((?!y).)*$', line):
        print(line)
```

... that has at least 2 consecutive vowels (a, e, i, o, u) like in the word "bear"

```python
for line in lines:
    if re.search(r'[aeiou]{2}', line):
        print(line)
```

... that has at least 3 vowels

```python
for line in lines:
    if re.search(r'[aeiou].*[aeiou].*[aeiou]', line):
        print(line)
```

... that has at least 30 characters

```python
for line in lines:
    if re.search(r'\w{30}', line):
        print(line)
```

... has the same word appear twice in the same line

```python
for line in lines:
    if re.search(r'', line):
        print(line)
```
