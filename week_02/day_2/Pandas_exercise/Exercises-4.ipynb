```python
import pandas as pd
```

### Download the data and load it to Pandas. 

You can find them [here](https://drive.google.com/file/d/1NY6cmF9Shjw-dD7BD6bNmfcIVz-kQcFR/view?usp=sharing).

```python
cast = pd.read_csv('cast.csv', index_col=None)
cast.head()
```

```python
release_dates = pd.read_csv('release_dates.csv', index_col=None, parse_dates=['date'], infer_datetime_format=True)
release_dates.head()
```

### Count the number of movies with "Christmas" in their title for each month, that is released are the USA.

```python
release_dates[(release_dates.title.str.contains('Christmas')) & (release_dates.country == 'USA')].date.dt.month.value_counts().sort_index()
```

### Count the number of movies with "Christmas" in their title for each month, that are released in Canada.

```python
release_dates[(release_dates.title.str.contains('Christmas')) & (release_dates.country == 'Canada')].date.dt.month.value_counts().sort_index()
```

### Count the number of movies in which the titles start with "The Hobbit" for each month, that are released in the USA.

```python
release_dates[(release_dates.title.str.startswith('The Hobbit')) & (release_dates.country == 'USA')].date.dt.month.value_counts().sort_index()
```

### Count the number of movies with "Romance" in their title for **each day of the week**, that are released in the USA.

```python
release_dates[(release_dates.title.str.contains('Romance')) & (release_dates.country == 'USA')].date.dt.dayofweek.value_counts().sort_index()
```

### Count the number of movies with "Action" in their title for **each day of the week**, that are released in the USA.

```python
release_dates[(release_dates.title.str.contains('Action')) & (release_dates.country == 'USA')].date.dt.dayofweek.value_counts().sort_index()
```

### On which date was each Judi Dench movie from the 1990s released in the USA?

```python
judi = cast[(cast.name == 'Judi Dench') & (cast.year // 10 == 199)]
release_dates[release_dates.country == 'USA'].merge(judi).sort_values('date')
```

### In which months do films with the actress Judi Dench tend to be released in the USA?

```python
judi = cast[cast.name == 'Judi Dench']
release_dates[release_dates.country == 'USA'].merge(judi).date.dt.month.value_counts().sort_values(ascending = False)
```

### In which months do films with the actor Tom Cruise tend to be released in the USA?

```python
tc = cast[cast.name == 'Tom Cruise']
release_dates[release_dates.country == 'USA'].merge(tc).date.dt.month.value_counts().sort_values(ascending = False)
```
